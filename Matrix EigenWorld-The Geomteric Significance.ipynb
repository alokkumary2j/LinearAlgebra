```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import linalg
```

```python
def computeAndPlotTransformedPoint(matrix,vectorX,vectorY):
    n=len(vectorX)
    transVectorX=np.ndarray(n)
    transVectorY=np.ndarray(n)
    fig = plt.figure()
    fig.suptitle(','.join(map(str,linalg.eig(matrix))))
    for i in np.arange(n):
        newVecStr=str(vectorX[i])+";"+str(vectorY[i])
        newVector=np.matrix(newVecStr)
        transfomedVect=np.dot(matrix,newVector)
        transVectorX[i]=transfomedVect.A[0][0]
        transVectorY[i]=transfomedVect.A[1][0]
    plt.scatter(vectorX,vectorY,marker='o',c='y')
    plt.scatter(transVectorX,transVectorY,marker='^',c='r')
    plt.show(True)
```

```python
n=100
m=10
vectorX=np.ndarray(n)
vectorY=np.ndarray(n)
index=0
for i in np.arange(m):
    for j in np.arange(m):
        vectorX[index]=i
        vectorY[index]=j
        index=index+1
```

```python
normalMatrices=[]
normalMatrices.append(np.matrix('15,0;0,8'))#EigenValues:15,8
normalMatrices.append(np.matrix('1,3;2,1'))#EigenValues:3,-1.4
normalMatrices.append(np.matrix('2,8;4,1'))#EigenValues:7,-4
normalMatrices.append(np.matrix('2,0;0,0'))#EigenValues:2,0
for i in np.arange(len(normalMatrices)):
    computeAndPlotTransformedPoint(normalMatrices[i],vectorX,vectorY)
```

```python
#Singular Matrices Look Innocent: But they DO REDUCE DIMENSIONS of Your VECTORS :-)
singularMatrices=[]
singularMatrices.append(np.matrix('1,1;1,1'))
singularMatrices.append(np.matrix('1,2;1,2'))
singularMatrices.append(np.matrix('1,2;2,4'))
singularMatrices.append(np.matrix('1,2;3,6'))
singularMatrices.append(np.matrix('4,6;12,18'))
for i in np.arange(len(singularMatrices)):
    computeAndPlotTransformedPoint(singularMatrices[i],vectorX,vectorY)
```

```python
almostSingularMatrices=[]
almostSingularMatrices.append(np.matrix('2,1;1.5,.85'))
for i in np.arange(len(almostSingularMatrices)):
    computeAndPlotTransformedPoint(almostSingularMatrices[i],vectorX,vectorY)
```
